#### 3) Tweet data
Coleta dos dados dos tweets presentes no arquivo twitter-archive-enhanced via API do Twitter.

```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True )
```

```python
t = time.process_time()
tweet_id = twitter_archive.tweet_id
tweet_data = []
tweet_error = []
query_error_count = 0
query_success_count = 0


for tweet in tweet_id:
    try:
        tweet_info = api.get_status(tweet, mode='extend')
        data = {'tweet_id':tweet, 'favorite_count':tweet_info.favorite_count, 'retweet_count':tweet_info.retweet_count}
        tweet_data.append(data)
        #print(tweet)
        query_success_count +=1
    except Exception as e:
        query_error_count +=1
        tweet_error.append(tweet)
        print(tweet, e)
        
            
print('Query Erros: ', query_error_count)
print('Query Success: ', query_success_count)

elapsed_time = time.process_time() - t
print(elapsed_time)  
```

```python
# Armazenando dados da consulta no arquivo tweet_json.txt
with open('tweet_json.txt', 'w') as file:
    for data in tweet_data:
        tweet_json = json.dumps(data)
        file.write(tweet_json+'\n')   
```

### Tabela image_predictions com 3 probabilidades

#### Definir

Entre as 3 probabilidades, escolher a melhor previsão.  

```python
image_predictions_clean[['p1', 'p1_conf', 'p1_dog', 'p2', 'p2_conf', 'p2_dog', 'p3', 'p3_conf', 'p3_dog']].head()
```

#### Programar

```python
# Mantendo apenas a melhor predição
dog = []
conf_dog = []

def probabilidade(df):
    ''' Verifica qual das 3 previsões (px) é verdadeira. Se mais de uma for verdadeira, 
        considera a de maior confiabilidade (px_conf).
        Argumentos:
            df: nome do dataframe
        Retorna: 
            dog =  Qual é o cachorro com a maior confiabilidade
            conf_dog = Confiabilidade da previsão     
    '''   
    if df.p1_dog == True:
        dog.append(df['p1'])
        conf_dog.append(df['p1_conf'])
    elif df.p2_dog == True:
        dog.append(df['p2'])
        conf_dog.append(df['p2_conf'])
    elif df.p3_dog == True:
        dog.append(df['p3'])
        conf_dog.append(df['p3_conf'])
    else:
        dog.append('no_dog')
        conf_dog.append(0)


image_predictions_clean.apply(probabilidade, axis=1)
       
image_predictions_clean['dog'] = dog
image_predictions_clean['conf_dog'] = conf_dog

```
